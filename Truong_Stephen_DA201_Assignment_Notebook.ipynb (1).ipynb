**Question 2:** Which service setting was the most popular for NHS North West London from 1 January to 1 June 2022?

```python
# For each of these service settings, determine the number of records available for the period and the location.Â 

# Service popularity with NHS North West London

nc_subset = nc.loc[(nc['appointment_date'] >= '01-01-2022') &
                    (nc['appointment_date'] <= '06-30-2022') & 
                    (nc['icb_ons_code']=='E54000027')].groupby(['service_setting','appointment_month']).sum()\
                    .sort_values(by=['count_of_appointments'],ascending=False)
                 
nc_subset

# Most popular service setting is General Practice.
```

**Question 3:** Which month had the highest number of appointments?

```python
# Number of appointments per month == sum of count_of_appointments by month.
# Use the groupby() and sort_values() functions.

nc_appt_highest = nc.groupby(['appointment_month'])['count_of_appointments'].agg('sum').reset_index()\
    .sort_values(by=['count_of_appointments'],ascending=False)

nc_appt_highest
```

**Question 4:** What was the total number of records per month?

```python
# Total number of records per month.

record_count = nc.groupby(['appointment_month'])['count_of_appointments'].agg('count').reset_index()\
    .sort_values(by=['count_of_appointments'],ascending=False)

record_count
```

**Question 1:** Should the NHS start looking at increasing staff levels? 

```python
# Create an aggregated data set to review the different features.

ar_agg_type = ar.loc[(ar["appointment_month"] >= '2021-08')
                     & (ar["appointment_month"] < '2024-08')].groupby('hcp_type')['count_of_appointments'].agg('sum').reset_index()

# View the DataFrame.

ar_agg_type
```

```python
# Determine the total number of appointments per month.

ar_test = ar.loc[(ar['appointment_month'] >= '2021-08') &
                    (ar['appointment_month'] < '2024-08')]\
                    .groupby(['appointment_month',
                              'hcp_type',
                              'time_between_book_and_appointment',
                              'appointment_status',
                              'appointment_mode']
                            ).agg(sum).reset_index()

# Add a new column to indicate the average utilisation of services.
# Monthly aggregate / 30 to get to a daily value.

ar_test['Utilisation'] = ar_test['count_of_appointments'] / 30

ar_test_sum = ar_test['Utilisation'].sum()

# View the DataFrame.

ar_test.round(decimals = 1 )
```

```python
# Plot sum of count of monthly visits.
# Convert the appointment_month to string data type for ease of visualisation.

ar_test['appointment_month'] = ar_test['appointment_month'].astype('string')

# Create a lineplot with Seaborn.

sns.lineplot( x = "appointment_month",
             y = "count_of_appointments",
             data = ar_test,
            ci=None);
```

```python
ar_agg_util = ar.loc[(ar["appointment_month"] >= '2021-08')
                     & (ar["appointment_month"] < '2024-08')].groupby('appointment_month')['count_of_appointments'].agg('sum').reset_index()

ar_agg_util['Utilisation'] = ar_agg_util['count_of_appointments'] / 30

ar_agg_util['%_Util'] = ar_agg_util['Utilisation'] / 1200000*100

ar_agg_util.round(decimals = 1 )
```

```python
# Plot monthly capacity utilisation.
# Create a lineplot.
# 12M max capacity per day

sns.lineplot( x = "appointment_month",
             y = "%_Util",
             data = ar_agg_util,
            ci=None);

# added an additional column that calculated the utilization based on the daily count. 
# NHS stated their daily max is 1,200,000 appointments, 
# divided the monthly by 30 and then used that value against 1,200,000 to determine the utilization on a monthly basis.
```
